# Generate segmentation masks

```python
import sys
import lightning as L
from omegaconf import OmegaConf
import matplotlib.pyplot as plt

sys.path.append("..")

from models import SegmentationModule
from data import LEM_Module
from utils import load_from_checkpoint, get_data_sample, time_series_to_rgb, get_preds_from_logits
```

```python
CONFIG_FILE_PATH = "../configs/train_config.yaml"

config = OmegaConf.load(CONFIG_FILE_PATH)
config
```

```python
model_name = SegmentationModule

model = model_name(config)
model = load_from_checkpoint(config, model, model_name=model_name)
```

```python
SAMPLE_INDICES = [5, 10, 15, 20, 30]

data_module = LEM_Module(config)
data_module.setup()
subset_loader = get_data_sample(data_module.test_dataset, indices=SAMPLE_INDICES)
```

```python
trainer = L.Trainer()
trainer.test(model, dataloaders=subset_loader)
outputs = trainer.predict(model, dataloaders=subset_loader)
```

```python
TIMESTAMP = 0

batch = outputs[0]
batch_inputs, batch_targets, batch_logits = batch["inputs"], batch["targets"], batch["logits"]
fig, axes = plt.subplots(nrows=3, ncols=len(SAMPLE_INDICES), figsize=(20, 12))

for i, (image, target, logits) in enumerate(zip(batch_inputs, batch_targets, batch_logits)):
    image = time_series_to_rgb(image)
    pred = get_preds_from_logits(logits, num_classes=config.model.num_classes)

    axes[0, i].imshow(image[TIMESTAMP]) # Plot image
    axes[1, i].imshow(target.squeeze(), cmap="binary", vmin=0, vmax=2) # Plot annotation
    axes[2, i].imshow(pred.squeeze(), cmap="binary", vmin=0, vmax=2) # Plot segmentation mask

plt.tight_layout()
plt.show()
```
